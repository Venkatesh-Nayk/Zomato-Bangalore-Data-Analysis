```python
import pandas as pd
```

```python
restaurant_details = pd.read_csv('Table_2_Final.csv')
restaurant_info = pd.read_csv('Table_1_Final.csv')
```

```python
restaurant_details.head()
```

```python
restaurant_info.head()
```

```python
restaurant_data = restaurant_details.merge(restaurant_info,left_index=True,right_index=True)
```

```python
restaurant_data.head()
```

```python
restaurant_data.drop(["Latitude","Longitude","id","hotel","link"],axis=1,inplace=True)
```

```python
restaurant_data
```

```python
def price(x):
    return int(x.split()[0][1:])

def place(x):
    return x.split(',')[-2].lower().strip()

def cus(x):
    return x.lower().strip()
```

```python
restaurant_data['Location'] = restaurant_data['location'].apply(lambda x : place(x))
restaurant_data['price'] = restaurant_data['price_for_one'].apply(lambda x : price(x))
```

```python
restaurant_data.rename(columns={"Location":"place"},inplace=True)
```

```python
restaurant_data
```

```python
restaurant_data.to_csv("Merged_data.csv",index=False)
```
